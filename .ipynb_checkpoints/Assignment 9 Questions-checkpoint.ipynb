```python
import sqlite3 as sql
import pandas as pd
import matplotlib.pyplot as plt
```

```python
conn = sql.connect('database.sqlite')
```

```python
def sql_query(q):
    return pd.read_sql_query(q,conn)
```

```python
q = 'SELECT * FROM SALARIES'

sql_query(q)
```

1. What is the average TotalPayBenefits for each JobTitle? exclude ('Not Provided', 'Not provided')

```python
q = ''' SELECT JobTitle, AVG(TotalPayBenefits) AS Average_Total_Pay_Benefits 
        FROM SALARIES
        WHERE JobTitle NOT IN ('Not Provided', 'Not provided')
        GROUP BY JobTitle;'''

sql_query(q)
```

2. Which JobTitle has the most employees in each year? exclude ('Not Provided', 'Not provided')

```python
q = """
SELECT Year, JobTitle, COUNT(*) AS EmployeeCount
FROM salaries
WHERE JobTitle NOT IN ('Not Provided', 'Not provided')
GROUP BY Year,JobTitle
ORDER BY Year, EmployeeCount DESC;
"""

all_results =  sql_query(q)
all_results
```

3. What is the total number of unique employees? exclude ('Not Provided', 'Not provided')

```python
q = '''SELECT COUNT(DISTINCT EmployeeName) AS TOTAL_UNIQUE_EMPLOYEE_COUNT
       FROM SALARIES
       WHERE EmployeeName NOT IN ('Not Provided', 'Not provided')'''

sql_query(q)
```

4. Show All Employees Ordered By Their TotalPayBenefits In Descending Order? exclude ('Not Provided', 'Not provided')

```python
q = '''SELECT EmployeeName,JobTitle, TotalPayBenefits
       FROM Salaries
       WHERE EmployeeName NOT IN ('Not Provided', 'Not provided')
       ORDER BY TotalPayBenefits DESC
       LIMIT 10'''

sql_query(q)
```

5. Show All Employees With A TotalPaybenefits Value Between 125,000 and 150,000 And A Job Title 'Firefighter'? exclude ('Not Provided', 'Not provided')

```python
q = '''SELECT EmployeeName, JobTitle, TotalPayBenefits
        FROM Salaries
        WHERE JobTitle == "Firefighter"
        AND TotalPayBenefits BETWEEN 125000 AND 150000 
        AND EmployeeName NOT IN ('Not Provided', 'Not provided')'''

sql_query(q)
```

6. How many employees have no overtime pay or other pay? exclude ('Not Provided', 'Not provided')


```python
q = '''SELECT COUNT(EmployeeName) AS Employees_with_no_overtimePay
        FROM Salaries
        WHERE OvertimePay == 0
        AND OtherPay == 0
        AND EmployeeName NOT IN ('Not Provided', 'Not provided')'''

sql_query(q)
```

7. Show top 10 employees having (overtimepay+OtherPay) in bar plot?  exclude ('Not Provided', 'Not provided')

```python
sql_query(q)
```

```python
q = '''SELECT EmployeeName, (OvertimePay+OtherPay) AS ExtraPay
        FROM Salaries
        WHERE EmployeeName NOT IN ('Not Provided', 'Not provided')
        ORDER BY ExtraPay DESC
        LIMIT 10'''

a = sql_query(q)
a.plot(kind='bar', x='EmployeeName', y='ExtraPay')
plt.xlabel('Employee Name')
plt.ylabel('Extra Pay')
plt.title('Top 10 Employee by Extra Pay (Overtime Pay + Other Pay)')
plt.show()
```

8. Show The Average of BasePay, OverTime, OtherPay for all employees in pie chart? exclude ('Not Provided', 'Not provided')


```python
q = '''SELECT AVG(BasePay) AS AverageBasePay,AVG(OvertimePay) AS AverageOvertimePay, AVG(OtherPay) AS AverageOtherPay
        FROM Salaries
        WHERE EmployeeName NOT IN ('Not Provided', 'Not provided')'''

a = sql_query(q)

values = a.iloc[0]  # Get the first row values
labels = a.columns   # Column names as labels

plt.figure(figsize=(8, 8))
plt.pie(values, labels=labels, autopct='%1.1f%%', colors=['lightblue', 'lightcoral', 'lightgreen'])
plt.title("Average Pay Distribution")
plt.show()
```

9. Get the average TotalPayBenefits for each JobTitle, but only for those with an average pay greater than 75000? exclude ('Not Provided', 'Not provided')


```python
q = '''SELECT JobTitle, TotalPay, AVG(TotalPayBenefits) AS AverageTotalPayBenefits
        FROM Salaries
        WHERE JobTitle NOT IN ('Not Provided', 'Not provided')
        GROUP BY JobTitle
        HAVING TotalPay > 75000'''

sql_query(q)
```
